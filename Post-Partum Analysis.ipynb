```python
# To calculate correlation between categorical variables
# Source for the function used: https://towardsdatascience.com/the-search-for-categorical-correlation-a1cf7f1888c9

import scipy.stats as ss

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x,y)
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
    rcorr = r-((r-1)**2)/(n-1)
    kcorr = k-((k-1)**2)/(n-1)

    return np.sqrt(phi2corr/min((kcorr-1),(rcorr-1)))
```

```python
corr = []
for col in columns:
  corr.append([col, cramers_v(df["feeling_anxious"], df[col])])
```

```python
df_corr = pd.DataFrame(corr, columns=["attribute", "correlation with feeling_anxious"])
df_corr.sort_values(by=["correlation with feeling_anxious"], axis=0, ascending=True, inplace=True)
```

# Data Pre-processing

```python
# Shuffling data before pre-processing

from sklearn.utils import shuffle

df = shuffle(df, random_state=49)
```

```python
X = df.drop(["feeling_anxious"], axis=1)
y = df["feeling_anxious"]
```

```python
# Ordinal encoding for age_group

age_groups = {'25-30': 1,
              '30-35': 2,
              '35-40': 3,
              '40-45': 4,
              '45-50': 5}
```

```python
X['age'] = X['age'].map(age_groups)
```

```python
# Dummy encoding for categorical variables

X = pd.get_dummies(data = X, columns = X.columns[1:])
```

```python
# Label Encoding for target variable

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# Splitting the data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=49)
```

# Modelling
We will be using 7 different classifiers and a Voting Classifier

```python
# Importing libraries for classification task

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
```

```python
# Initialization

logistic = LogisticRegression(max_iter=1000, random_state=49)
support = SVC()
naive = GaussianNB()
decision = DecisionTreeClassifier(max_depth=5, min_samples_leaf=0.01, min_samples_split=2, random_state=49)
random = RandomForestClassifier(max_depth=6, min_samples_leaf=0.01, min_samples_split=2, n_estimators=100, random_state=49)
xgb = XGBClassifier(max_depth=6, learning_rate=0.01, random_state=49)
mlp = MLPClassifier(solver='adam', alpha=1e-4, hidden_layer_sizes=(40,40,), max_iter=1000, random_state=49)

voting = VotingClassifier(
    estimators=[
        ('logistic', logistic),
        ('support', support),
        ('naive', naive),
        ('decision', decision),
        ('random', random),
        ('ada', xgb),
        ('mlp', mlp)
    ], voting="hard"
)
```

```python
# Setting up the evaluation metrics

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

evaluation = pd.DataFrame(columns=["Model", "Accuracy", "Precision", "Recall", "F1-Score", "ROC-AUC Score"])

def evaluate(name, model):
  model.fit(X_train, y_train)

  y_pred = model.predict(X_test)

  accuracy = np.round(accuracy_score(y_test, y_pred)*100, 2)
  precision = np.round(precision_score(y_test, y_pred)*100, 2)
  recall = np.round(recall_score(y_test, y_pred)*100, 2)
  f1 = np.round(f1_score(y_test, y_pred)*100, 2)
  roc_auc = np.round(roc_auc_score(y_test, y_pred)*100, 2)

  evaluation.loc[len(evaluation)] = [name, accuracy, precision, recall, f1, roc_auc]
```

```python
# Training the models

evaluate("Logistic", logistic)
evaluate("Support Vector Machine", support)
evaluate("Gaussian Naive-Bayes", naive)
evaluate("Decision Tree", decision)
evaluate("Random Forest", random)
evaluate("Xtreme Gradient Boost", xgb)
evaluate("Multi-Layer Perceptron", mlp)
evaluate("Voting Classifier", voting)
```

```python
# Determining and Creating a Confusion Matrix

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

mlp.fit(X_train, y_train)
y_pred = mlp.predict(X_test)

c_matrix = confusion_matrix(y_test, y_pred)
display = ConfusionMatrixDisplay(c_matrix, display_labels=mlp.classes_)
display.plot()
```
